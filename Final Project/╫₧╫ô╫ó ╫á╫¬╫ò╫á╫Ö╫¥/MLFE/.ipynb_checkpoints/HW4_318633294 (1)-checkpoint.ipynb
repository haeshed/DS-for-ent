````python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
#     s = np.array([xx1.ravel(), xx2.ravel()]).T ###
#     print(s.shape) ####
#     return ###
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
````

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

````python
class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.theta = None
        self.costs = []
    
    def sigmoid(self, X, theta):
        sig =  1 / (1 + np.exp(-(X.dot(theta))))
        return sig

        
    
    def compute_cost(self, X, y, theta): # computing the loss in data tiwh the given X, y and thetas
        prob = self.sigmoid(X,theta)
        step1 = - y * (np.log(prob))
        step2 = (1 - y) * (np.log(1 - prob))
        step3 = step1 - step2
        loss = (1 / len(X)) * np.sum(step3)
        return loss
    


    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        ones = np.ones(len(X)) #bias trick
        X = np.column_stack((ones, X))
        np.random.seed(self.random_state)
        theta = np.random.random(len(X[0])) # new theta with random weights and the same size of X
        theta_ = theta.copy() # avoid changing the original thetas
       
        for num in range(self.n_iter):
            cost = self.compute_cost(X, y, theta)
            if num > 0 and self.costs[-1] - cost < self.eps:
                break
            cost = self.compute_cost(X, y, theta)
            self.costs.insert(num, cost)
            for j in range(len(theta)):
                theta[j] -=  self.eta * (self.sigmoid(X,theta) - y).dot(X[:, j])
        self.theta = theta  


    def predict(self,X):
        #bias trick
        ones = np.ones(len(X)) 
        X = np.column_stack((ones, X))  
        predicts = np.array([1 if self.sigmoid(X[ind], self.theta) > 0.5 else 0 for ind in range(len(X))])
        return predicts
    
    def accuracy(self, X, y):
        predicts = self.predict(X)
        correct_predict = np.count_nonzero(predicts == y)
        accuracy = correct_predict / len(y)
        return accuracy

````

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

````python

training_set = pd.read_csv('training_set.csv')
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]

shuffle_training = training_set.values
np.random.shuffle(shuffle_training) #shuffle the traning set
final_accuracy = 0
best_accuracy = 0
divide_five = int(len(shuffle_training) / 5)# compute the length of each fold
size = 0
for eta in etas:
    for eps in epss:
        validation_accuracy = 0
        start = 0
        end = divide_five
        for num in range(5):# split the shuffle_training into five folds and compute the accuracy with different validation set each time
            X_validation = shuffle_training[start:end, :-1]  
            y_validation = shuffle_training[start:end, -1]
            X_new_train = np.concatenate((shuffle_training[0:start, :-1], shuffle_training[end: , :-1]), axis = 0)
            y_new_train = np.concatenate((shuffle_training[0:start, -1], shuffle_training[end: , -1]), axis = 0)
            LoR = LogisticRegressionGD(eta = eta, eps = eps)
            LoR.fit(X_new_train, y_new_train)
            validation_accuracy += LoR.accuracy(X_validation, y_validation)
            start = end
            end += divide_five
            
                
        size = 1    
        final_accuracy = validation_accuracy / 5
        if final_accuracy > best_accuracy:
            best_eta = eta
            best_eps = eps
            best_accuracy = final_accuracy
            

print(f"best eta is {best_eta}, best_eps is {best_eps}")
print(f"the best_accuracy is {best_accuracy * 100}")
````

# Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

````python
#### Your code here ####
# calc normal pdf    

def norm_pdf(data, mu, sigma):
    step1 = 1 / (sigma * np.sqrt(2 * np.pi))
    step2 = np.exp(-0.5 * ((data - mu) / sigma) ** 2)
    step3 = step1 * step2
    return step3
    
    
````

### Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

````python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k = 1, n_iter=1000, eps=0.01, random_state = 42):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps
        self.costs = []
        self.random_state = random_state

    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        
        """
        np.random.seed(self.random_state)
        self.weights = np.full(self.k, (1 / self.k))
        self.mus = np.random.random(self.k)
        self.sigmas = np.random.random(self.k)
        
    
    
    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        step1 = np.array([self.weights[k] * norm_pdf(data, self.mus[k], self.sigmas[k]) for k in range(self.k)])
        step2  = step1.sum(axis = 0)
        responsibilities = step1 / step2
        self.responsibilities = responsibilities.T
        
        
    def maximization(self, data):
        """
        M step - updating distribution params
        """
        self.weights = self.responsibilities.mean(axis = 0)
        for j in range(self.k):
            self.mus[j] = self.responsibilities[:, j].dot(data) / (len(data) * self.weights[j])
            self.sigmas[j] = np.sqrt(self.responsibilities[:,j].dot((data - self.mus[j]) ** 2) / (len(data) * self.weights[j]))

    
    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then q and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.
        """
        self.init_params(data)
        for num in range(self.n_iter): 
            cost = self.compute_cost(data)
            if num > 0 and self.costs[-1] - cost < self.eps:
                break
            self.costs.insert(num, cost)
            self.expectation(data)
            self.maximization(data)
            
    
    def compute_cost(self, data):
        sum_cost = 0
        for j in range (self.k):
            sum_cost += np.sum(- np.log2(self.weights[j] * norm_pdf(data, self.mus[j], self.sigmas[j])))
        return sum_cost    
                               
    
    def get_dist_params(self):
        params = np.array([self.weights, self.mus, self.sigmas]).T
        return params
    
    
````

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


````python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1, random_state = 42):
        self.k = k
        self.random_state = random_state
        self.classes = []
        self.priors = []
        self.parameters = []
        
        
    def class_update(self, y):
        labels, occurrences = np.unique(y, return_counts=True)
        self.classes = labels.tolist()
        for i in range(len(self.classes)):
            self.priors = np.append(self.priors, occurrences[i] / len(y))
        

    def fit(self, X, y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        self.class_update(y)
        for ind in range(len(self.classes)):
            class_data = X[y == self.classes[ind]]
            self.parameters.append([])
            for fea in range(len(X[0])):
                em = EM(k = self.k, random_state = self.random_state)
                em.fit(class_data[:, fea])
                feature_params = em.get_dist_params()
                self.parameters[ind].append(feature_params)
                
    def get_data_likelihoods(self, X, label): 
        """ 
        X : data matrix
        label: class value
        return : entire data likelihood
        """
        label_index = self.classes.index(label)
        data_liklihood = 1
        for feature in range(len(X[0])):
            feature_likelihood = 0
            for j in range(self.k):
                weight_j = self.parameters[label_index][feature][j][0]
                mu_j = self.parameters[label_index][feature][j][1]
                sig_j = self.parameters[label_index][feature][j][2]
                feature_likelihood += norm_pdf(X[:, feature], mu_j, sig_j) * weight_j 
            data_liklihood *= feature_likelihood
            
        return data_liklihood
      

    def get_all_data_posterior(self, X, label):
        """ 
        X : data matrix
        label: class value
        return : entire data posterior
        """
        label_index = self.classes.index(label)
        posterior =  self.get_data_likelihoods(X, label) * self.priors[label_index]
        return posterior

 
    def predict(self, X): # X is a matrix 
        """
        Return the predicted class label"""
        all_posteriors = np.array([[self.get_all_data_posterior(X, label)] for label in self.classes])
        all_posteriors = all_posteriors.T
        label_predicts = np.array([self.classes[np.argmax(row)] for row in all_posteriors])
        return label_predicts
    
    def accuracy(self, X, y):
        label_predicts = self.predict(X)
        correct_predict = np.count_nonzero(label_predicts == y)
        accuracy = correct_predict / len(y)
        return accuracy
    
        
    
    
        
````
