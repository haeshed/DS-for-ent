````python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
````

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

````python
class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.J_history = []
        self.theta = None

    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        #Add column of ones for bias trick
        X = np.concatenate((X, np.ones((X.shape[0], 1))), axis=1)
        
        np.random.seed(self.random_state)
        
        #Initialize a random theta for the logistic regression
        self.theta = np.random.rand(X.shape[1])
        
        #Fit the training data
        while self.n_iter > 0:
            self.n_iter -= 1
            
            #Calculate the gradient in each iteration and update theta
            gradient = np.dot(self.sigmoid(X) - y, X) / y.shape[0]
            self.theta = self.theta - self.eta * gradient
            
            #Compute the cost and check if we have reached convergence
            self.J_history.append(self.compute_cost(X, y))
            if len(self.J_history) > 1 and self.J_history[-2] - self.J_history[-1] < self.eps:
                break
        
    def compute_cost(self, X, y):
        #Compute the logistic regression cost function
        sig = self.sigmoid(X)
        J = (np.dot(-y, np.log(sig)) - np.dot((1 - y), np.log(1 - sig))) / y.shape[0]
        return J
        
    def sigmoid(self, X):
        #Calculate the sigmoid function for the relevant data
        expo = np.dot(X, self.theta)
        return 1 / (1 + np.exp(-expo))
    
    def predict(self, X):        
        """Return the predicted class label"""
        #Add column of ones for bias trick
        X = np.concatenate((X, np.ones((X.shape[0], 1))), axis=1)
        
        #Return the class label with the highest probability 
        return np.round(self.sigmoid(X))
        
````

## Cross Validation

(5 points)

Use 5-fold cross validation in order to find the best eps and eta params from the given lists.

Shuffle the training set before you split the data to the folds.

````python
#### Your code here ####
etas = [0.05, 0.005, 0.0005, 0.00005, 0.000005]
epss = [0.01, 0.001, 0.0001, 0.00001, 0.000001]

#Declare variables for the best values
best_eta, best_eps ,best_acc = None, None, 0

#Create new dataframe for better visualization of the results
accuracies = pd.DataFrame(index=etas, columns=epss)

np.random.seed(42)
#Create array with the training set indices, shuffle them and create new shuffled training data
shuffled_indices = np.arange(X_training.shape[0])
np.random.shuffle(shuffled_indices)
X_shuffled = X_training[shuffled_indices]
y_shuffled = y_training[shuffled_indices]

#Split the shuffled data into 5 folds
X_split = np.split(X_shuffled, 5)
y_split = np.split(y_shuffled, 5)

#Perform 5-fold cross validation to find the best eps and eta
for eta in etas:
    for eps in epss:
        accuracy = np.zeros(5)
        for i in range(5):
            #Create copy of the splitted data
            X_split_copy = X_split.copy()
            y_split_copy = y_split.copy()
            
            #Choose one of the splitted arrays to use as validation data
            X_fold_valid = X_split[i]
            y_fold_valid = y_split[i]
            
            #Remove the validation data from the splitted array
            del X_split_copy[i]
            del y_split_copy[i]
            
            #Concatenate the splitted data back without the validation data
            X_fold_train = np.concatenate(X_split_copy)
            y_fold_train = np.concatenate(y_split_copy)
            
            #Create new logistic regression model and train it on the current fold data
            LoR = LogisticRegressionGD(eta=eta, eps=eps)
            LoR.fit(X_fold_train, y_fold_train)
            
            #Calculate predictions and accuracy for the current fold
            predictions = LoR.predict(X_fold_valid)
            accuracy[i] = np.sum(1*(predictions == y_fold_valid)) / y_fold_valid.shape[0] 
        #Calculate the avarage accuracy for the current eta and eps and locate in the dataframe
        avg_acc = np.sum(accuracy) / 5
        accuracies.loc[eta, eps] = avg_acc
        
        #Update the new best accuracy, eps and eta if necessary
        if(avg_acc > best_acc):
            best_acc = avg_acc
            best_eta = eta
            best_eps = eps
            
#Present the dataframe and best eps and eta values
print("Vertical values are for eta and horizontal are for epsilon")
display(accuracies)
print("Best accuracy is", best_acc, "and it is for eta =", best_eta, "and eps =", best_eps)
        
        
````

## Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

````python
#### Your code here ####
# calc normal pdf    
def norm_pdf(data, mu, sigma):   
    #Calculte PDF in parts according to the normal distribution fucnctiom.
    pi = (2 * np.pi) ** 0.5   
    expo = -0.5 * ((np.repeat(data[:,None], mu.shape[0], axis=1) - mu) / sigma) ** 2
    return np.exp(expo) / (pi * sigma)
````

## Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

````python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1, n_iter=1000, eps=0.01):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps

    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        """
        #Assign inital values for the distribution parameters
        self.mu = (np.arange(self.k)+ 1) * np.mean(data) / self.k
        self.sigma = (np.arange(self.k)+ 1) * np.std(data) / self.k
        self.w = np.ones(self.k) / self.k

    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        #Calculate responsibilities for all instances for each Guassian and update the class field
        normal_dist = norm_pdf(data, self.mu, self.sigma)
        likelihood = normal_dist * (np.atleast_1d(self.w)[:,None]).T
        likelihood_sum = likelihood.sum(axis=1)[:,None]
        self.respon = likelihood / likelihood_sum
        
        #Return the following to help calculating the cost later
        return likelihood_sum

    def maximization(self, data):
        """
        M step - updating distribution params
        """
        #Update w with the new responsibilities
        w = np.mean(self.respon, axis=0)
        
        #Update mu and sigma using the new w
        mu = (self.respon.T.dot(data[:,None]) / (w[:,None] * data.shape[0])).ravel()
        sigma = np.sqrt(np.mean(np.multiply(self.respon, np.square(np.array(data)[:, np.newaxis] - mu)), axis=0) / w)
        
        return w, mu, sigma

    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then expectation and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.
        """
        #Initialize parameters and cost according to the data
        self.init_params(data)
        prev_cost = -np.inf
        
        #Fit the training data
        while self.n_iter > 1:
            self.n_iter -= 1
            
            #Expectation part
            likelihood = self.expectation(data)
            
            #Compute cost and check if the difference is small enough for assuming convergence
            curr_cost = -np.sum(np.log(np.sum(likelihood ,axis=1)))
            if np.abs(prev_cost - curr_cost) < self.eps:      
                break
            prev_cost = curr_cost
            
            #Maximization part
            self.w, self.mu, self.sigma = self.maximization(data)
            
                
    def compute_cost(self, likelihood):
        return np.sum(-np.log(likelihood))       

    def get_dist_params(self):
        return self.mu, self.sigma, self.w
    
````

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


````python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=1):
        self.k = k

    def fit(self, X, y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        #Get numbers of instances with each class value
        self.labels, counts = np.unique(y, return_counts=True)
        
        #Initialize class fields for the prior probabilities and a list of EM objects
        # to train the data for each class label and feature
        self.prior = np.zeros(self.labels.shape[0])
        self.EMs = []
        
        #Iterate over all class labels
        for i in range (self.labels.shape[0]):
            self.EMs.append([])
            
            #Calculate prior probability for each class label
            self.prior[i] = counts[i] / y.shape[0]
            
            #Extract relevant data for each class label
            label_data = X[self.labels[i] == y]
            
            #Train an EM object for each feature
            for j in range(X.shape[1]):
                self.EMs[i].append(EM(k=self.k))
                self.EMs[i][-1].fit(label_data[:,j])
             
    def predict(self, X):
        """Return the predicted class label"""
        
        #Get likelihoods for each class label for each feature for all of the samples
        likelihoods = np.zeros((self.labels.shape[0], X.shape[1], X.shape[0]))
        
        #Iterate over all class labels
        for i in range (likelihoods.shape[0]):
            #Iterate over all features for the given class label
            for j in range(X.shape[1]):
                #Extract the relevant feature data
                feature_data = X[:, j]
                likelihoods[i,j] = np.max(self.EMs[i][j].expectation(feature_data), axis=1)
                
        #Calculate posterior probability for each class label
        posteriors = np.zeros((self.labels.shape[0], X.shape[0]))                     
        for i in range (self.labels.shape[0]):
            #For every instance, multiply all lilkelihoods and the prior probability
            posteriors[i] = self.prior[i] * np.prod(likelihoods[i], axis=0)
            
        #Predict the class label with the highest posterior probability
        predictions = np.argmax(posteriors, axis=0)
        return predictions         
````

**Question 1B)**

The following dataset will work with logistic regression better than naive bayes because we have created a dataset which is not conditionally independent. <br>
Naive bayes assumes conditional independence between all of the features.
This means that given the class label, the instances features are independent.
First, we have generated 500 instances that are strongly dependent.
We did it by using a covariance matrix that has high values outside the main diagonal which implies high dependence between the features.<br>
The covariance matrix also has low values on the main diagonal which indicates low variance on each feature.<br>
This means that the data is supposed to be densed in a small range of values. 
We gave all of those instances class label zero.<br>
We repeated the steps for 500 more instances that are dependent, and then united both datasets into one big dataset with both class labels.<br>
By this implementation we try to achieve strong conditional dependence.<br>
By all said above, we can deduce that Logistic Regression will work better than Naive Bayes on the generated data.  


````python
mean = [[1,5,10],[30,35,40]]
cov = [[[0.5,7,9],[7,0.7,4],[9,4,0.6]], [[0.3,4,11],[4,0.2,15],[11,15,0.6]]]
LoR_data_class0 = np.random.multivariate_normal(mean[0], cov[0], 500)
LoR_label_class0 = np.zeros(500)
LoR_data_class1 = np.random.multivariate_normal(mean[1], cov[1], 500)
LoR_label_class1 = np.ones(500)

LoR_data = np.concatenate((LoR_data_class0, LoR_data_class1))
LoR_labels = np.concatenate((LoR_label_class0, LoR_label_class1))
print("First 20 instances of the data:\n", LoR_data[:20])
````
