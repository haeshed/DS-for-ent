## Part 1- Reading the data & basic statistics

```python
# A2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
# A3
data = pd.read_csv('Mobileye_risk_TelAviv_collision_likelihood_.csv')
```

```python
# A4
data.describe()
```

```python
# A5 - Proportion of classification feature
low_risk,high_risk=data['collision_likelihood'].value_counts(normalize=True)*100
low_risk=float("{0:.2f}".format(low_risk))
high_risk=float("{0:.2f}".format(high_risk))
print("The proportion are: \n Safe roads: ",low_risk," %" "\n Dangerous roads: ",high_risk," %")
```

```python
# A6
data.isna().sum()
```

## Part 2 - Preporocessing

```python
# A7
data.drop('harsh_braking_ratio', inplace=True, axis=1)
```

```python
# A8. a,b
data['near_miss_pedestrian_ratio'].fillna(data['near_miss_pedestrian_ratio'].mean(), inplace=True)
data['near_miss_bicycle_ratio'].fillna(data['near_miss_bicycle_ratio'].median(), inplace=True)
```

```python
# A8.c. Function that replaces NaN values with normal dist. random. 
def fill_NaN_with_normalrand(data_column):
    values = data_column.values
    mask = np.isnan(values) # mask of NaNs
    mu, sigma = data_column.mean(), data_column.std()
    values[mask] = np.random.normal(mu, sigma, size=mask.sum())
    return data_column

# A8.d. The function is replacing the NaN values 
fill_NaN_with_normalrand(data['braking_count'])
```

```python
# A8.e. 
data.isnull().sum()
```

```python
# A 9-10
X = data.drop('collision_likelihood', axis=1)
y = data.collision_likelihood

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

## Part 3 - Decision Tree

### Part 3A - Building an optimal tree model with Accuracy metric

```python
# A 11-13
train_acc = []
test_acc = []

for d in range(1,21):
    tree = DecisionTreeClassifier(max_depth=d, min_samples_leaf=10, random_state=0)
    tree.fit(X_train, y_train)
    train_acc.append(tree.score(X_train, y_train))
    test_acc.append(tree.score(X_test, y_test))
```

```python
# A 14
plt.style.use('dark_background')
fig = plt.figure()
ax1 = fig.add_subplot(111)


ax1.scatter(range(1, 21), train_acc, s=30, marker="s", label='Training Accuracy')
ax1.scatter(range(1, 21) ,test_acc, s=30, marker="o", label='Testing Accuracy')
plt.plot(range(1, 21),train_acc)
plt.plot(range(1, 21),test_acc)
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Max Depth")
plt.legend(loc='upper left');
plt.show()
```
